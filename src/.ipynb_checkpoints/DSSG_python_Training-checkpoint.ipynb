```python
import re  #regular expressions library
```

```python
def spaces_to_snake(column_name): 
    """
    converts a string that has spaces into snake_case
        Example: print camel_to_snale("KENNY BROUGHT HIS WIFE")
        > KENNY_BROUGHT_HIS_WIFE
        
    To see how to apply this to camel case, see:
     http://stackoverflow.com/questions/1175208/elegant-python-function-to-convert-camelcase-to-camel-case
    """
    s = re.sub(r"\s+", '_', column_name)   #changes spaces to _
    return s.lower()      #returns lower case
```

```python
df.columns = [spaces_to_snake(col) for col in df.columns]  #apply spaces_to_snake to columns for all columns in df.columns

#Note: a much more elegant and pythonic way to do this would be to use the rename method and lambda syntax 
#(e.g. df.rename(columns=lambda x: x.strip())), but we'll cover this later.
        
```
